### Basic Math Functions: Min and Max
So we have found the sum, what if I want to know the minimum and maximum number ranking that I gave the different animals?

We can use the `min()` and `max()` functions just like we used the `sum()` function. 

It doesn't look like the ranking column has any missing data, so we don't need to use `na.rm=TRUE` this time.

In the cell below, find the minumum and maximum ranking I gave out.

```r
# Find the minimum ranking of the stuffed animals in the dataset
min(TestData$Ranking, na.rm = TRUE)
# Find the maximum ranking of the stuffed animals in the dataset
max(TestData$Ranking, na.rm = TRUE)
```

Okay, those are the min and max rankings, but **WHO GOT THEM? WHICH ANIMAL?** 

This is the trickiest bit of code for the day. Hang tight, and if it doesn't click for you right away that is okay. This is one I usually chat with about students 1-on-1 to clear up questions. 

We are going to use the `which.` or the plain `which` function to find entire rows within our dataframe and extract them. 

Your code will ultimately look like this:  
`TestData[which.max(TestData$Ranking),]`  
or  
`TestData[which(TestData$Ranking == max(TestData$Ranking)),]`  

The first option will **only show the first instance of a maximum. If you have two entries with the same maximum value, it will only show one.**  


The second option will show **all instances of a maximum**. <font color="teal"> **THIS IS THE BETTER WAY TO FIND MAX/MIN VALUES AND YOU SHOULD DO THIS WAY IN YOUR ASSIGNMENTS**.</font>

So why does the code look the way it does in the two options?   

Let's start with the first one:`TestData[which.max(TestData$Ranking),]`  
 
> `TestData[,]` says that we are going to extract some data from the `TestData` dataframe. Inside the brackets, you put `row number,column number`. If you leave one or more blank, it takes the **entire row or column**.

>For example, `TestData[2,3]` would extract a data point that is in the second row, third column.  
Similarly, `TestData[2,]` would extract the entire second row since we didn't specify a column.

> So back to our example: `TestData[which.max(TestData$Ranking),]`. In the row number position, we have `which.max(TestData$Ranking)`. This tells the computer to find which place has the maximum value of the `Ranking` column and put it in that place. But remember, it only takes the first instance. You can use `which.min` the same way you would use `which.max`.

Now we can look at the second method: `TestData[which(TestData$Ranking == max(TestData$Ranking)),]`  (THIS IS THE BETTER METHOD!!)
> in this syntax, the only thing that has changed is the way our `which` function is being used in the row number place. 

>Here, instead of using the built-in `which.max` that will stop at the first instance, we tell it to find which rows have data where the `TestData$Ranking` column is equivalent (`==`) to the `max()` value of the `TestData$Ranking` column.

**I KNOW THIS IS A LOT.** Take a deep breath, remember the joy of writing nice clean markdown cells and rest assured that writing in R will soon be a pleasant experience too.

In the cell below, extract the rows containing the minimum and maximum values of the `Ranking` column **using both syntax options for each min and max**. They will produce the same output for `min`, but will produce different outputs for `max`.

```r
# Extract the minimum value(s) of the Ranking column using Method 1
TestData[which.min(TestData$Ranking),]
# Extract the maximum value(s) of the Ranking column using Method 1
TestData[which.max(TestData$Ranking),]
# Extract the minimum value of the Ranking column using Method 2
TestData[which(TestData$Ranking == min(TestData$Ranking)),]
# Extract the maximum value of the Ranking column using Method 2
TestData[which(TestData$Ranking == max(TestData$Ranking)),]
```
