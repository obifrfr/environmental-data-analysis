## What is the biggest occurrence of a tsunami?

Quite possibly the most significant occurence of a tsunami to date, the [Great Alaska Tsunami of 1964](https://www.usgs.gov/natural-hazards/tsunami/1964-great-alaska-earthquake-and-tsunami) triggered a massive undersea earthquake that caused widespread devastation along the Alakan coastline and beyond.  

<img src="https://earthquake.alaska.edu/sites/default/files/PacificAnim1964.gif" alt="image link" width="500"/>

* With a magnitude of **9.2**, the quake generated a tsunami with heights of over 67 meters (220 feet) in some areas including [Shoup Bay, Valdez Inlet, and along the Alaskan Gulf Coast](https://www.mycg.uscg.mil/Portals/6/1_%20Quake%20chart%20%28NOAA%29.jpg).
* These immense waves caused [extensive damage to coastal communities, infrastructure, and ecosystems across Alaska](https://www.ngdc.noaa.gov/hazel/view/hazards/tsunami/event-more-info/106).

## What is the worst instance of a tsunami?

The worst instance of a tsunami in recorded history is the 2004 Indian Ocean tsunami, also known as the [Boxing Day tsunami](https://www.ngdc.noaa.gov/hazel/view/hazards/tsunami/event-more-info/179).

<img src="https://www.worldvision.org/wp-content/uploads/2018/12/D170-0020-087_61087.jpg" alt="image link" width="500"/>

* The devastating event was triggered by a massive undersea earthquake with a magnitude of *9.1-9.3* off the coast of [Sumatra, Indonesia](https://www.researchgate.net/publication/339915010/figure/fig2/AS:868680846032898@1584121061706/Map-of-Indonesia-and-Sumatra-Island-Figure-3-Location-of-Mukomuko-City-In-Sumatra-Island.jpg). 
* The tsunami waves reached heights of up to 30 meters (100 feet) and affected numerous countries bordering the Indian Ocean, causing widespread destruction and loss of life.
* More than **230,000** people lost their lives in this catastrophe, making it one of the deadliest natural disasters in history.

## How common are tsunamis?

Tsunamis are [relatively infrequent](https://www.tsunami.gov/tsuabout) compared to other natural disasters, but they can have devastating consequences when they do occur. They are most commonly generated by undersea earthquakes, but can also be triggered by volcanic eruptions, landslides, or meteorite impacts.

The frequency of tsunamis varies [depending on the region and geological activity](https://www.usgs.gov/natural-hazards/tsunami). Some areas, such as the [Pacific Ring of Fire](https://media.istockphoto.com/id/1307701663/vector/vector-map-of-the-pacific-ring-of-fire-with-the-main-volcanoes.jpg?s=612x612&w=0&k=20&c=fu3RQ9AcjWLT5GpuGyGZIy2Pltse3dbIzTU_rkPBPEU=), experience more frequent tsunamis due to the high levels of tectonic activity in the region. Other regions may experience tsunamis less frequently.

<img src="https://cat.ingv.it/images/tsunami-world-map.jpeg" alt="image link" width="500"/>

> Global map of tsunami distribution.

While tsunamis are not as common as other natural disasters like earthquakes or hurricanes, their potential for widespread devastation underscores the importance of preparedness, education, and proactive measures to mitigate their impact on vulnerable coastal regions.

## Where do tsunamis occur? 

[Tsunamis](https://www.nationalgeographic.org/encyclopedia/tsunami/) predominantly occur in the Pacific Ocean, particularly along the **_Pacific Ring of Fire_** , which is characterized by high levels of tectonic activity. This region, encircling the Pacific Ocean basin, is home to numerous undersea earthquakes, volcanic eruptions, and other geological phenomena that can trigger tsunamis.

<img src="https://media.istockphoto.com/id/1307701663/vector/vector-map-of-the-pacific-ring-of-fire-with-the-main-volcanoes.jpg?s=612x612&w=0&k=20&c=fu3RQ9AcjWLT5GpuGyGZIy2Pltse3dbIzTU_rkPBPEU=" alt="image link" width="500"/>

Coastal regions near [subduction zones](https://d9-wret.s3.us-west-2.amazonaws.com/assets/palladium/production/s3fs-public/styles/full_width/public/thumbnails/image/subzon.gif?itok=-KhqdFzI), where one tectonic plate is forced beneath another, are particularly susceptible to tsunamis. These subduction zones have the potential to generate *large* and *destructive* tsunamis.


## Are tsunamis preventable?

Prevention of tsunamis themselves is not feasible due to their natural causes. However, efforts focused on [early warning systems](https://ioc.unesco.org/tsunami-early-warning-systems), [disaster preparedness](https://www.tsunami.gov/tsupreparedness), and [coastal planning](https://nws.weather.gov/nthmp/documents/designingfortsunamis.pdf) can significantly reduce the loss of life and property damage associated with tsunamis.

<img src="https://www.japantimes.co.jp/uploads/imported_images/uploads/2023/01/np_file_203593.jpeg" alt="image link" width="500"/>

> Concrete seawalls developed by seismic researchers in Japan.

**Coastal engineering projects**, such as the construction of tsunami barriers or natural vegetation buffers, can help mitigate the impact of tsunamis by reducing wave energy and protecting coastal infrastructure. Additionally, public education and community preparedness initiatives play a crucial role in raising awareness about tsunamis and ensuring that individuals and communities [know how to respond](https://www.nationalgeographic.com/science/article/how-to-survive-a-tsunami) in the event of a tsunami warning.


## Can you mitigate the damage of tsunamis?

While it is challenging to prevent tsunamis entirely, there are various measures that can be implemented to [mitigate their damage](https://www.unisdr.org/tsunamis) and protect coastal communities.

Implementing robust [early warning systems](https://ioc.unesco.org/tsunami-early-warning-systems) can provide timely alerts to at-risk populations, allowing them to evacuate to higher ground before the arrival of a tsunami. These systems utilize technologies such as seismometers, tsunami detection buoys, and communication networks to monitor seismic activity and issue warnings.

<img src="https://www.tsunami.noaa.gov/sites/default/files/styles/landing_page_banner/public/thumbnails/image/dart-2.jpg?itok=3SG1RLsv" alt="image link" width="500"/>

Employing [natural and nature-based solutions](https://www.tsunami.gov/tsupreparedness), such as **mangrove forests** can act as natural barriers to reduce wave energy and protect coastal communities from tsunami impacts. These ecosystems also provide additional benefits such as biodiversity conservation and shoreline stabilization.  


<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/b/b1/Mangrove_ecosystem_in_the_coastal_intertidal_zone.png/440px-Mangrove_ecosystem_in_the_coastal_intertidal_zone.png" alt="image link" width="500"/>

Constructing [tsunami-resistant structures](https://www.tsunami.gov/tsupreparedness), such as seawalls can help minimize the impact of tsunamis on coastal infrastructure. These structures are designed to absorb and dissipate wave energy, reducing flooding and erosion.

By implementing a combination of these [mitigation measures](https://www.unisdr.org/tsunamis), it is possible to reduce the damage and loss of life associated with tsunamis.

## Can you recover/remediate damage from tsunamis after the fact?

While tsunamis *can* cause significant damage to coastal communities and infrastructure, recovery and remediation efforts can be undertaken after the event to address the damage and restore affected areas.

<img src="https://idsb.tmgrup.com.tr/ly/uploads/images/2021/03/04/97557.jpg" alt="image link" width="500"/>

> Remediation efforts in Japan: On the left, the immediate effects of the 2011 Tohoku disaster; On the right, the same road in 2021.

* Following a tsunami, efforts are made to repair or reconstruct damaged infrastructure such as buildings, roads, bridges, and utilities. This involves assessing the extent of damage, prioritizing repair projects, and mobilizing resources for [reconstruction](https://www.worldbank.org/en/results/2013/04/01/tsunami-recovery-and-reconstruction).

* Tsunamis can have detrimental effects on the environment, including habitat destruction, pollution, and coastal erosion. [Remediation efforts](https://www.worldbank.org/en/results/2013/04/01/tsunami-recovery-and-reconstruction) may include restoring ecosystems, cleaning up debris and hazardous materials, and implementing measures to prevent further environmental degradation.

* Recovery efforts extend beyond physical infrastructure to include [rebuilding communities](https://www.redcross.org/about-us/news-and-events/news/2019/tsunami-recovery.html) and enhancing resilience. This involves providing support for displaced populations, facilitating community-led recovery initiatives, and implementing measures to reduce vulnerability to future disasters.

## Helpful tsunami resources

- [NOAA Tsunami Website](https://www.tsunami.gov/): The official website of the National Oceanic and Atmospheric Administration (NOAA) provides comprehensive information on tsunamis, including educational resources, real-time tsunami monitoring, and preparedness tips.

- [Pacific Tsunami Warning Center](https://ptwc.weather.gov/): Operated by NOAA, the Pacific Tsunami Warning Center monitors seismic activity in the Pacific Ocean region and issues tsunami warnings and advisories to at-risk populations.

- [United Nations Office for Disaster Risk Reduction (UNDRR) - Tsunamis](https://www.undrr.org/): UNDRR offers resources and publications on tsunami risk reduction, resilience-building initiatives, and best practices for disaster preparedness and response.

- [USGS Earthquake Hazards Program - Tsunamis](https://www.usgs.gov/natural-hazards/tsunami): The U.S. Geological Survey (USGS) provides valuable information on tsunami science, monitoring, research, and preparedness efforts.

- [American Red Cross - Tsunami Preparedness](https://www.redcross.org/get-help/how-to-prepare-for-emergencies/types-of-emergencies/tsunami.html): The American Red Cross offers practical guidance and resources on tsunami preparedness, evacuation planning, and emergency response.

- [National Geographic - Tsunamis](https://www.nationalgeographic.com/environment/natural-disasters/tsunamis/): National Geographic features in-depth articles, videos, and interactive resources exploring the science, history, and impact of tsunamis around the world.

- [TsunamiReady Program](https://www.weather.gov/tsunamiready/): The TsunamiReady program, administered by the National Weather Service, recognizes communities and organizations that have taken proactive steps to prepare for and respond to tsunamis.

- [UNESCO Intergovernmental Oceanographic Commission - Tsunami Program](https://ioc.unesco.org/tsunami): UNESCO's Tsunami Program promotes international cooperation in tsunami research, monitoring, and capacity-building efforts to enhance tsunami resilience worldwide.
