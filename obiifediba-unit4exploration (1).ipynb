## ggplot2 package - the ultimate plotting package
We are going to use a package called ggplot2 to make most of our data visualizations. 

The "gg" in "ggplot" stands for grammar of graphics. This was designed to make plotting all different kinds of plot follow the same structure. So if you can get comfortable making scatter plots, for example, then making histograms or contours or GIS plots or any other plot type will be essentially the same commands and just as easy.

The guide is broken out into plot types - one variable continuous, 2 variable continuous, bivariate distributions, etc. 

**At the top of each category is the base command you have to type in to make the plotting start. THIS IS YOUR FIRST STEP AND YOU MUST ALWAYS DO IT**

For single variable plots, it is `ggplot(data = DataframeName, mapping = aes(x=ColumnNameOnly))`. Note that this can also be written more concise, or "tidy," as `ggplot(DataframeName,aes(ColumnNameOnly))`.

For two variable plots (like scatter plots), you would use `ggplot(data = DataframeName, mapping = aes(x=ColumnName1, y=ColumnName2))` or `ggplot(DataframeName,aes(ColumnName1,ColumnName2))`

The `aes` in here stands for "aesthetic," indicating that these are the columns of data to be used in making your visual, or aesthetic, creation.

**After you set up your plots with the ggplot(,aes()) command, then start adding "layers" of information/formatting**.
You start doing this by adding a `+` after you finish you `ggplot(,aes())` command, then on a new line you can add more information, such as what type of plot you actually want to make. 

***

If you wanted to make a scatterplot of Ranking versus Stuffed Animal Count from that TestData dataframe we used last week about my opinions on various animals, you would write the following code:
```
ggplot(data = TestData, aes(x = Ranking, y=StuffedAnimalNumber) +
    geom_point()
```

**OR**

```
ggplot(data=TestData)+
    geom_point(mapping=aes(x=Ranking, y=StuffedAnimalNumber))
```

The first variation is good for when you want to use the same x and y data for all of your plot layers, the second variation is good for if you are overlaying plots that maybe share an x axis but have different y values.

## Now that we've got our questions in mind, 
# Time to import (*and clean*) some data!

````r
#Import data
tsudata <- read.csv("https://raw.githubusercontent.com/emilyfairfax/ESRM203/master/TSUcsv.csv")
tsudata <- tsudata[!is.na(tsudata$MaxWaterHeight) & !is.na(tsudata$WarningStatus),]

#Check that it worked
print("tsunami mami")
````

## Summary Statistics

````r
#Basic summary stats
summary(tsudata)
````

# EDA Question 1
## Is there a trend in the **_severity_** of tsunamis over the years, as indicated by variables like *MaxWaterHeight*?
* Climate deniers may not like this question. But it must be asked, are tsunamis becoming increasingly more severe?
* The answer to this question can be observed by observing the serverity, measured in height as the years go on.

````r
#Exploratory plot 1: Tsunami Severity Over Time
ggplot(tsudata, aes(x = Year, y = MaxWaterHeight)) + 
    geom_bar(stat = "summary", fun = "mean") + 
    labs(title = "Trend in Tsunami Severity Over Time in Years",
        x = "Year",
        y = "Max Water Height") +
    theme_minimal()
````

## What can we deduce from this analysis?
 * Since the earliest recording of a tsunami, severity (as observed by the MaxWaterHeight variable) has drastically increased, especially since 2000.
 * There is an overall lack of data between the span of 1980 to the mid-to-late 1990's.
 * The highest recorded height occured in 2005.

# EDA Question 2
## What are the most **_common warning statuses_** (*WarningStatus*) issued before tsunamis occur?
* Warning status can be an important precursor to an effective tsunami evacuation response.
* Timely warning statuses can save lives, preventing unneccesary casualities and allowing civilans to get to safety.

````r
#Exploratory plot 2: Most common warning statuses issued before tsunamis occur
ggplot(tsudata, aes(x = WarningStatus)) +
    geom_bar() +
    labs(title = "Most Common Warning Statuses Before Tsunamis",
        x = "Warning Status",
        y = "Frequency") +
    theme_minimal()
````

## What can we deduce from this analysis?
 * For starters, higher warning statuses are likely to implicit a more rapid and effective evacuation response than lower ones.
 * This phenomenon could explain why a majority of warning statuses are released at the Level 4: the *highest* severity.

# EDA Question 3
## How does the *IntensitySoloviev* variable correlate with the **_number of deaths_** (*DeathCount*) caused by tsunamis?
* The Soloviev scale is used as a measure of tsunami "size". By utilizing this scale, one can effectively determine how the size of a tsunami can affect its death toll.

````r
# Plotting IntensitySoloviev against DeathCount
ggplot(tsudata, aes(x = IntensitySoloviev, y = DeathCount)) +
  geom_point() +
  labs(title = "Correlation between Intensity and Death Count",
       x = "Intensity Soloviev",
       y = "Death Count")
````

## What can we deduce (**_if anything_**) from this analysis?
 * While it can be reasonably deduced that the "larger" a natural disaster is, the greater its death toll will be, this data is lacking in sufficient evidence.
 * Because of a lack of Soloviev scale observations in the dataset, a reasonable conclusion was **NOT** able to be made.
 * Therefore, a further analysis must be conducted in the future to observe this relationship.
